```python
import sys
import matplotlib.pyplot as plt
import healpy as hp
from pathlib import Path

sys.path.append("..")

from cmbstack.main import StackingPipeline
```

### Load theoretical spectrum

```python
DATA_DIR = Path('../data')

fname = 'base_plikHM_TTTEEE_lowl_lowE_lensing.minimum.theory_cl'
```

### Run Peak Stacking Pipeline

```python
pipeline = StackingPipeline.from_cl(DATA_DIR / fname, 1024, 42)
pipeline.run()
```

### Simulated CMB temperature map

```python
plt.figure()
hp.mollview(pipeline.map,title=None,unit=r'$\mu$k')
```

### Mean stacked peak

```python
plt.figure(figsize=(6, 5))
plt.imshow(pipeline.stacked,origin='lower')
plt.xlabel('Pixel x')
plt.ylabel('Pixel y')
plt.colorbar(label=r'Temperature ($\sigma$ units)')
plt.show()
```

### Radial profile

```python
plt.figure(figsize=(6,5))
plt.plot(pipeline.radius,pipeline.profile)
plt.xlabel('Radius (arcmin)')
plt.ylabel(r'Temperature profile ($\sigma$ units)')
plt.show()
```
